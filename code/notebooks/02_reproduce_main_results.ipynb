```python
import os
from pathlib import Path
import pandas as pd

ROOT = Path.cwd()
DATA = Path(os.environ.get("NEURIPS_DATA_DIR", ROOT / "data"))
if not DATA.is_absolute():
    DATA = ROOT / DATA
exposure = pd.read_csv(DATA / "core" / "nation_exposure_enriched.csv")
print(f"National exposure countries: {len(exposure)}")
print("Exposure range:", exposure["weighted_exposure"].min(), "to", exposure["weighted_exposure"].max())
display(exposure.sort_values("weighted_exposure", ascending=False)[["country_code", "country_name", "weighted_exposure"]].head(10))
display(exposure.sort_values("weighted_exposure")[["country_code", "country_name", "weighted_exposure"]].head(10))
```

```python
gender = pd.read_csv(DATA / "mechanisms" / "gender_gap.csv")
reliable_gender = gender[gender["reliable"].astype(bool)].copy()
female_higher = (reliable_gender["female_exposure"] > reliable_gender["male_exposure"]).sum()
median_gap = reliable_gender["relative_gender_gap"].median()
print(f"Reliable gender comparisons: {len(reliable_gender)}")
print(f"Female exposure higher: {female_higher}/{len(reliable_gender)}")
print(f"Median relative gender gap: {median_gap:.3%}")
display(pd.read_csv(DATA / "mechanisms" / "gender_decomposition_summary.csv"))
```

```python
adoption = pd.read_csv(DATA / "validation" / "adoption_predictor_grid.csv")
pivot = adoption.pivot(index="outcome_label", columns="predictor_set", values="r2")
pivot[["Exposure", "WC + log GNI", "WC + log GNI + Exposure"]]
```
